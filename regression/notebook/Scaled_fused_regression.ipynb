# **Prepare data**

```python
import pandas as pd
import numpy as np

df = pd.read_csv('new_fused_dataset.csv')
df.head()
```

```python
#check if there are missing values
df.isnull().sum()

#check the data frame info
df.info()

#calculate the descriptive statistics
df.describe()
```

# **One-hot encoding for 'status' column**


*   Status_developing = 1 (developing)
*   Status_developing = 0 (developed)



```python
df = pd.get_dummies(df, columns=['Status'], drop_first=True, dtype=int)
#check the first 10 rows
print(df.head(10))

#check the data frame after one hot encoding
print(df.info())
```

# **Select Feature**
â€œHow severe is the suicide rate in this country?â€

```python
x = df.drop('suicides/100k pop', axis=1)
y = df['suicides/100k pop']
```

#A. **80-20 Train test split**

```python
from sklearn.model_selection import train_test_split

x_80_train, x_80_test, y_80_train, y_80_test = train_test_split(
    x, y, test_size=0.2, random_state=42
    )

print('Shape of X_train:', x_80_train.shape)
print('Shape of X_test:', x_80_test.shape)
print('Shape of y_train:', y_80_train.shape)
print('Shape of y_test:', y_80_test.shape)
```

# **Feature Scaling (80-20 split)**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_80_train_scaled = scaler.fit_transform(x_80_train)
x_80_test_scaled = scaler.transform(x_80_test)

print('Shape of scaled X_train:', x_80_train_scaled.shape)
print('Shape of scaled X_test:', x_80_test_scaled.shape)
```

# **Model Training (with 80-20 scaled data)**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import pandas as pd
import numpy as np

models = {
    'Random Forest': RandomForestRegressor(random_state=42),
    'Linear Regression': LinearRegression(),
    'K-Nearest Neighbors': KNeighborsRegressor(),
    'Support Vector Machine': SVR(),
    'Neural Network': MLPRegressor(max_iter=1000, random_state=42) # Increased max_iter for convergence
}

results_scaled_80_20 = []

for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(x_80_train_scaled, y_80_train)
    y_pred = model.predict(x_80_test_scaled)

    #evaluate
    rmse = np.sqrt(mean_squared_error(y_80_test, y_pred))
    mae = mean_absolute_error(y_80_test, y_pred)

    results_scaled_80_20.append({
        'Model': name,
        'RMSE': rmse,
        'MAE': mae
    })

results_scaled_80_20_df = pd.DataFrame(results_scaled_80_20)
print("\nModel Performance (80-20 Split, Scaled Features):")
display(results_scaled_80_20_df)
```

#B. **70-30 Train test split**

```python

x_70_train, x_70_test, y_70_train, y_70_test = train_test_split(
    x, y, test_size=0.3, random_state=42
    )

print('Shape of X_train:', x_70_train.shape)
print('Shape of X_test:', x_70_test.shape)
print('Shape of y_train:', y_70_train.shape)
print('Shape of y_test:', y_70_test.shape)
```

# **Feature Scaling (70-30 split)**

```python
from sklearn.preprocessing import StandardScaler

scaler_70_30 = StandardScaler()
x_70_train_scaled = scaler_70_30.fit_transform(x_70_train)
x_70_test_scaled = scaler_70_30.transform(x_70_test)

print('Shape of scaled X_train (70-30 split):', x_70_train_scaled.shape)
print('Shape of scaled X_test (70-30 split):', x_70_test_scaled.shape)
```

# **Model Training (with 70-30 scaled data)**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import pandas as pd
import numpy as np

models_70_30 = {
    'Random Forest': RandomForestRegressor(random_state=42),
    'Linear Regression': LinearRegression(),
    'K-Nearest Neighbors': KNeighborsRegressor(),
    'Support Vector Machine': SVR(),
    'Neural Network': MLPRegressor(max_iter=1000, random_state=42) # Increased max_iter for convergence
}

results_scaled_70_30 = []

for name, model in models_70_30.items():
    print(f"\nTraining {name}...")
    model.fit(x_70_train_scaled, y_70_train)
    y_pred_70_30 = model.predict(x_70_test_scaled)

    #evaluate
    rmse_70_30 = np.sqrt(mean_squared_error(y_70_test, y_pred_70_30))
    mae_70_30 = mean_absolute_error(y_70_test, y_pred_70_30)

    results_scaled_70_30.append({
        'Model': name,
        'RMSE': rmse_70_30,
        'MAE': mae_70_30
    })

results_scaled_70_30_df = pd.DataFrame(results_scaled_70_30)
print("\nModel Performance (70-30 Split, Scaled Features):")
display(results_scaled_70_30_df)
```

# **Hyperparameter Tuning (80-20 Split)**

# **Hyperparameter Tuning with GridSearchCV (80-20 Split)**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Random Forest Regressor Grid Search
param_grid_rf = {
    'n_estimators': [50, 100, 150],
    'max_depth': [10, 20, None],
    'min_samples_leaf': [1, 2, 4]
}
rf_grid = GridSearchCV(RandomForestRegressor(random_state=42), param_grid_rf, cv=3, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)
rf_grid.fit(x_80_train_scaled, y_80_train)
best_rf_model_grid = rf_grid.best_estimator_
print("Best Random Forest (Grid Search) Parameters:", rf_grid.best_params_)
```

```python
from sklearn.neural_network import MLPRegressor

# Neural Network Regressor Grid Search
param_grid_mlp = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'activation': ['relu', 'tanh'],
    'alpha': [0.0001, 0.001]
}
mlp_grid = GridSearchCV(MLPRegressor(max_iter=1000, random_state=42), param_grid_mlp, cv=3, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)
mlp_grid.fit(x_80_train_scaled, y_80_train)
best_mlp_model_grid = mlp_grid.best_estimator_
print("Best Neural Network (Grid Search) Parameters:", mlp_grid.best_params_)
```

# **Hyperparameter Tuning with RandomizedSearchCV (80-20 Split)**

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform

# Random Forest Regressor Randomized Search
param_dist_rf = {
    'n_estimators': randint(50, 200),
    'max_depth': [10, 20, 30, None],
    'min_samples_leaf': randint(1, 5)
}
rf_random = RandomizedSearchCV(RandomForestRegressor(random_state=42), param_distributions=param_dist_rf, n_iter=10, cv=3, scoring='neg_mean_squared_error', n_jobs=-1, random_state=42, verbose=1)
rf_random.fit(x_80_train_scaled, y_80_train)
best_rf_model_random = rf_random.best_estimator_
print("Best Random Forest (Randomized Search) Parameters:", rf_random.best_params_)
```

```python
from sklearn.neural_network import MLPRegressor

# Neural Network Regressor Randomized Search
param_dist_mlp = {
    'hidden_layer_sizes': [(randint.rvs(50, 150),), (randint.rvs(50, 100), randint.rvs(20, 70))],
    'activation': ['relu', 'tanh'],
    'alpha': uniform(0.00001, 0.001)
}
mlp_random = RandomizedSearchCV(MLPRegressor(max_iter=1000, random_state=42), param_distributions=param_dist_mlp, n_iter=10, cv=3, scoring='neg_mean_squared_error', n_jobs=-1, random_state=42, verbose=1)
mlp_random.fit(x_80_train_scaled, y_80_train)
best_mlp_model_random = mlp_random.best_estimator_
print("Best Neural Network (Randomized Search) Parameters:", mlp_random.best_params_)
```

# **Evaluation**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd

results = []

# Evaluate Best Random Forest Model (from GridSearchCV)
y_pred_rf_grid = best_rf_model_grid.predict(x_80_test_scaled)
rmse_rf_grid = np.sqrt(mean_squared_error(y_80_test, y_pred_rf_grid))
mae_rf_grid = mean_absolute_error(y_80_test, y_pred_rf_grid)
results.append({'Model': 'Random Forest (Grid)', 'RMSE': rmse_rf_grid, 'MAE': mae_rf_grid})

# Evaluate Best Neural Network Model (from GridSearchCV)
y_pred_mlp_grid = best_mlp_model_grid.predict(x_80_test_scaled)
rmse_mlp_grid = np.sqrt(mean_squared_error(y_80_test, y_pred_mlp_grid))
mae_mlp_grid = mean_absolute_error(y_80_test, y_pred_mlp_grid)
results.append({'Model': 'Neural Network (Grid)', 'RMSE': rmse_mlp_grid, 'MAE': mae_mlp_grid})

# Evaluate Best Random Forest Model (from Randomized Search)
y_pred_rf_random = best_rf_model_random.predict(x_80_test_scaled)
rmse_rf_random = np.sqrt(mean_squared_error(y_80_test, y_pred_rf_random))
mae_rf_random = mean_absolute_error(y_80_test, y_pred_rf_random)
results.append({'Model': 'Random Forest (Randomized)', 'RMSE': rmse_rf_random, 'MAE': mae_rf_random})

# Evaluate Best Neural Network Model (from Randomized Search)
y_pred_mlp_random = best_mlp_model_random.predict(x_80_test_scaled)
rmse_mlp_random = np.sqrt(mean_squared_error(y_80_test, y_pred_mlp_random))
mae_mlp_random = mean_absolute_error(y_80_test, y_pred_mlp_random)
results.append({'Model': 'Neural Network (Randomized)', 'RMSE': rmse_mlp_random, 'MAE': mae_mlp_random})

# Create a DataFrame to display results
tuned_results = pd.DataFrame(results)
print("\nPerformance of Tuned Models on 80-20 Test Set (Grid vs. Randomized Search):")
display(tuned_results)
```

# **XAI EXECUTION**

```python
# =====================================
# XAI (SHAP) â€“ Regression (ANN, Fused Dataset)
# =====================================

import shap
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------
# Select the final model to explain
# -------------------------------------
final_model = best_mlp_model_grid   # <- CONSISTENT with your evaluation

# Feature names from original dataframe
feature_names = x.columns.tolist()

# -------------------------------------
# Background data (small subset)
# -------------------------------------
np.random.seed(42)
background = x_80_test_scaled[
    np.random.choice(x_80_test_scaled.shape[0], 100, replace=False)
]

# -------------------------------------
# Create SHAP explainer (REGRESSION)
# -------------------------------------
explainer = shap.KernelExplainer(
    final_model.predict,
    background
)

# -------------------------------------
# Compute SHAP values
# -------------------------------------
X_explain = x_80_test_scaled[:50]
shap_values = explainer.shap_values(X_explain)

# -------------------------------------
# Global Feature Importance (Bar Plot)
# -------------------------------------
shap.summary_plot(
    shap_values,
    X_explain,
    feature_names=feature_names,
    plot_type="bar"
)

# -------------------------------------
# SHAP Summary Plot (Impact + Direction)
# -------------------------------------
shap.summary_plot(
    shap_values,
    X_explain,
    feature_names=feature_names
)

# -------------------------------------
# Local Explanation (Single Instance)
# -------------------------------------
sample_index = 0

shap.plots.waterfall(
    shap.Explanation(
        values=shap_values[sample_index],
        base_values=explainer.expected_value,
        data=X_explain[sample_index],
        feature_names=feature_names
    )
)
```

Now that AutoGluon is installed, we can use it to perform AutoML regression. We will use the `TabularPredictor` for this task, which automatically handles preprocessing, model selection, and hyperparameter tuning.

We'll use an 80-20 train-test split for this, similar to your previous manual modeling approach. AutoGluon's `fit` method will handle the training and internal validation.

```python
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.model_selection import train_test_split
import pandas as pd

# Ensure df_pycaret is defined, if not, reconstruct it from x and y
if 'df_pycaret' not in locals():
    df_pycaret = pd.concat([x, y], axis=1)

# Split the data into training and test sets (80-20 split)
train_data, test_data = train_test_split(df_pycaret, test_size=0.2, random_state=42)

# Initialize TabularPredictor
predictor = TabularPredictor(label='suicides/100k pop', eval_metric='rmse', path='AutogluonModels').fit(train_data, presets='best_quality')

# Make predictions on the test data
y_pred_autogluon = predictor.predict(test_data)

# Evaluate the model
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

rmse_autogluon = np.sqrt(mean_squared_error(test_data['suicides/100k pop'], y_pred_autogluon))
mae_autogluon = mean_absolute_error(test_data['suicides/100k pop'], y_pred_autogluon)

print(f"\nAutoGluon Model Performance (80-20 Split):")
print(f"RMSE: {rmse_autogluon:.4f}")
print(f"MAE: {mae_autogluon:.4f}")
```
