# **Prepare data**

```python
import pandas as pd
import numpy as np

df = pd.read_csv('new_fused_dataset.csv')
df.head()
```

```python
#check if there are missing values
df.isnull().sum()

#check the data frame info
df.info()

#calculate the descriptive statistics
df.describe()
```

# **One-hot encoding for 'status' column**


*   Status_developing = 1 (developing)
*   Status_developing = 0 (developed)



```python
df = pd.get_dummies(df, columns=['Status'], drop_first=True, dtype=int)
#check the first 10 rows
print(df.head(10))

#check the data frame after one hot encoding
print(df.info())
```

# **Select Feature**
“How severe is the suicide rate in this country?”

```python
x = df.drop('suicides/100k pop', axis=1)
y = df['suicides/100k pop']
```

#A. **80-20 Train test split**

```python
from sklearn.model_selection import train_test_split

x_80_train, x_80_test, y_80_train, y_80_test = train_test_split(
    x, y, test_size=0.2, random_state=42
    )

print('Shape of X_train:', x_80_train.shape)
print('Shape of X_test:', x_80_test.shape)
print('Shape of y_train:', y_80_train.shape)
print('Shape of y_test:', y_80_test.shape)
```

# **Model Training (without scaling)**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

models = {
    'Random Forest': RandomForestRegressor(),
    'Linear Regression': LinearRegression(),
    'K-Nearest Neighbors': KNeighborsRegressor(),
    'Support Vector Machine': SVR(),
    'Neural Network': MLPRegressor()
}

results = []

for name, model in models.items():
    model.fit(x_80_train, y_80_train)
    y_pred = model.predict(x_80_test)

    #evaluate
    rmse = np.sqrt(mean_squared_error(y_80_test, y_pred))
    mae = mean_absolute_error(y_80_test, y_pred)



    results.append({
        'Model': name,
        'RMSE': rmse,
        'MAE': mae

    })

results_df = pd.DataFrame(results)
print(results_df)
```
