#**GRAPH 3: Hyperparameter Tuning Comparison



FIRST DATASET

```python
import matplotlib.pyplot as plt
import numpy as np# Algorithms
algorithms_d1 = ["Random Forest", "k-NN"]

# RMSE values
grid_rmse_d1 = [18.2517, 18.5676]
random_rmse_d1 = [18.4668, 18.6291]

x = np.arange(len(algorithms_d1))
width = 0.35

plt.figure()
plt.bar(x - width/2, grid_rmse_d1, width, label="GridSearch")
plt.bar(x + width/2, random_rmse_d1, width, label="RandomSearch")

plt.xticks(x, algorithms_d1)
plt.ylabel("RMSE")
plt.title("Hyperparameter Tuning Comparison (D1: Suicide Dataset)")
plt.legend()
plt.show()
```
